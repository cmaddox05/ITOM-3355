## Part 3: Types of Testing

### Closed-Box Testing

**Definition**: Test without looking at the code

**Approach**: Treat function as a "black box"
- Input goes in
- Output comes out
- Don't care HOW it works, just that it does

**Example**: Testing a calculator
- Input: 5 + 3
- Expected output: 8
- Don't need to know the algorithm

### Open-Box Testing

**Definition**: Test while looking at the code

**Approach**: Read the code to find potential issues
- Look for loops (test first/last element)
- Look for conditionals (test each branch)
- Look for calculations (test boundary values)

**We'll use BOTH!**

## Part 4: Doctest - Python's Built-in Testing Tool

### What is Doctest?

A way to write tests directly in your docstring!

### Format:
```python
def function_name(params):
    '''
    Description of function.
    
    >>> function_name(test_input)
    expected_output
    >>> function_name(another_input)
    another_output
    '''
```

**Key Points:**
- `>>>` indicates a test (like Python prompt)
- Next line is expected result
- NO quotes around expected result
- Exact match required!

### Example with Doctest:

````python
def add_numbers(a, b):
    '''
    Return the sum of a and b.
    
    >>> add_numbers(5, 3)
    8
    >>> add_numbers(10, 20)
    30
    >>> add_numbers(-5, 5)
    0
    '''
    return a + b

# Run the tests!
import doctest
doctest.testmod(verbose=True)
````

### How to Run Doctest:

Add this at the bottom of your cell:
```python
import doctest
doctest.testmod(verbose=True)
```

**Output:**
- âœ… All tests pass: You'll see "0 failures"
- âŒ Test fails: You'll see which test failed and why

## Part 5: Common Use Cases vs Edge Cases

### Common Use Cases

**Definition**: Typical, expected inputs

**Examples:**
- Normal numbers (5, 10, 100)
- Typical strings ("hello", "Python")
- Standard lists ([1, 2, 3, 4])

### Edge Cases

**Definition**: Unusual but possible inputs that might break code

**Examples:**
- Empty list: `[]`
- Zero: `0`
- Negative numbers: `-5`
- Empty string: `""`
- Single item: `[42]`
- Very large numbers
- Boundaries (first/last in loop)

### Why Test Edge Cases?

Most bugs happen at the edges! If code works for edge cases, it usually works for everything.

## Part 6: Complete Example - The Right Way

Let's design a function the professional way:

````python
def larger(num1, num2):
    '''
    Return the larger of two numbers.
    
    Common cases:
    >>> larger(5, 3)
    5
    >>> larger(10, 20)
    20
    
    Edge cases:
    >>> larger(5, 5)
    5
    >>> larger(-10, -5)
    -5
    >>> larger(0, 0)
    0
    '''
    # Let Copilot write this, then test!
    if num1 > num2:
        return num1
    else:
        return num2

import doctest
doctest.testmod(verbose=True)
````

````python
def divide(a, b):
    '''
    Return a divided by b.
    
    This test might fail due to floating point precision:
    >>> divide(10, 3)
    3.333333333333333...
    
    Better: Round in your test
    >>> round(divide(10, 3), 2)
    3.33
    '''
    return a / b

import doctest
doctest.testmod(verbose=True)
````

## Part 10: When Tests Fail - Debugging

### What To Do When Tests Fail:

1. **Read the error message carefully**
   - What test failed?
   - What was expected vs what you got?

2. **Check your test first**
   - Is the expected output correct?
   - Did you format it right?

3. **Look at the code**
   - Trace through with the failing input
   - Use print statements to debug

4. **Fix and retest**
   - Make the change
   - Run tests again
   - Repeat until all pass

### Example of Debugging:

````python
def is_passing_grade(score):
    '''
    Return True if score is 60 or above, False otherwise.
    
    >>> is_passing_grade(70)
    True
    >>> is_passing_grade(59)
    False
    >>> is_passing_grade(60)
    True
    '''
    # WRONG CODE (on purpose):
    return score > 60  # Should be >=

# This will fail! The test at 60 will fail.
# Fix: Change > to >=

import doctest
doctest.testmod(verbose=True)
````
