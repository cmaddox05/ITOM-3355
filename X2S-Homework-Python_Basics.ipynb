## Problem 1: Personal Finance Calculator (20 points)

Create a program that helps someone plan their monthly budget.

**Requirements**:
1. Ask the user for:
   - Monthly income (as float)
   - Rent/mortgage payment (as float)
   - Utilities cost (as float)
   - Food budget (as float)
   - Transportation cost (as float)

2. Calculate:
   - Total expenses
   - Remaining money (income - expenses)
   - Each expense as a percentage of income (rounded to 1 decimal)

3. Display:
   - Each expense with its percentage
   - Total expenses
   - Remaining money
   - A message: "Under budget" if remaining is positive, "Over budget" if negative

**Example Output**:
```
Monthly Income: $4000.00

Expenses:
- Rent: $1200.00 (30.0%)
- Utilities: $150.00 (3.8%)
- Food: $400.00 (10.0%)
- Transportation: $250.00 (6.3%)

Total Expenses: $2000.00
Remaining: $2000.00
Status: Under budget
```

````python
#Create a float from user input for variables Monthly Income, Rent/Mortgage Payment, Utilities Cost, Food Cost, and Transportation Cost
monthly_income = float(input("Enter your monthly income: "))
rent_mortgage = float(input("Enter your monthly rent/mortgage payment: "))
utilities = float(input("Enter your monthly utilities cost: "))
food = float(input("Enter your monthly food cost: "))
transportation = float(input("Enter your monthly transportation cost: "))

print(f"Monthly Income: ${monthly_income:.2f}")

#Print individual expenses followed by what percentage of the monthly income each expense represents rounded to one decimal place
print(f"Rent/Mortgage Payment: ${rent_mortgage:.2f} ({(rent_mortgage / monthly_income) * 100:.1f}%)")
print(f"Utilities Cost: ${utilities:.2f} ({(utilities / monthly_income) * 100:.1f}%)")
print(f"Food Cost: ${food:.2f} ({(food / monthly_income) * 100:.1f}%)")
print(f"Transportation Cost: ${transportation:.2f} ({(transportation / monthly_income) * 100:.1f}%)")
total_expenses = rent_mortgage + utilities + food + transportation
print(f"Monthly Expenses: ${total_expenses:.2f}")

remaining_balance = monthly_income - total_expenses
print(f"Remaining Balance: ${remaining_balance:.2f}")

if remaining_balance < 0:
    print("Over budget")
else:
    print("Within budget")
````

## Problem 2: Product Inventory Manager (25 points)

Create a program that manages a small product inventory.

**Requirements**:

1. Create three lists:
   - `product_names` = ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam']
   - `prices` = [899.99, 25.50, 75.00, 299.99, 89.99]

````python
#Your code here
product_names = ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam']
product_prices = [999.99, 25.50, 75.00, 299.99, 89.99]
quantities = [5, 15, 8 , 3 , 12]

#Print a formatted table displaying the product name, price, and quantity in stock labelled "Product Inventory"
print("Product Inventory")
print("-" * 40)
for i in range(len(product_names)):
    print(f"{product_names[i]:<12} ${product_prices[i]:<8} {quantities[i]}")

#create an analysis section that calculates and displays total inventory value, most expensive product,  least expensive product, average price, low stock products (less than 1 items), and low stock products(which products)
total_inventory_value = sum(price * quantity for price, quantity in zip(product_prices, quantities))
most_expensive_product = product_names[product_prices.index(max(product_prices))]
most_expensive_price = max(product_prices)
#create another variable for most expensive and least expensive product price as well
    
least_expensive_product = product_names[product_prices.index(min(product_prices))]
least_expensive_price = min(product_prices)
average_price = sum(product_prices) / len(product_prices)
low_stock_products = [product_names[i] for i in range(len(quantities)) if quantities[i] < 5]

print(f"\nTotal Inventory Value: ${total_inventory_value:.2f}")
print(f"Most Expensive Product: {most_expensive_product}, ${most_expensive_price:.2f}")
print(f"Least Expensive Product: {least_expensive_product}, ${least_expensive_price:.2f}")
print(f"Average Price: ${average_price:.2f}")
#make an if statement to print out a low stock message if there are low stock products
if low_stock_products:  print(f"Low Stock Product Alert: {len(low_stock_products)} Products")

#print the low stock products
for product in low_stock_products:
    print(f"Low Stock Items: {product}")
````

## Problem 3: Sales Commission Calculator (20 points)

Create a program that calculates a salesperson's commission based on tiered rates.

**Commission Structure**:
- First $1000 in sales: 5% commission
- Next $2000 in sales ($1001-$3000): 7.5% commission
- All sales above $3000: 10% commission

**Requirements**:
1. Ask user for total sales amount
2. Calculate commission using the tiered structure
3. Display:
   - Total sales
   - Commission breakdown by tier
   - Total commission earned
   - Total payout (if base salary is $2000)

**Example Output for $4500 in sales**:
```
Sales Report
============
Total Sales: $4500.00

Commission Breakdown:
- First $1000 at 5%: $50.00
- Next $2000 at 7.5%: $150.00
- Remaining $1500 at 10%: $150.00

Total Commission: $350.00
Base Salary: $2000.00
Total Payout: $2350.00
```

**Hint**: You'll need to use conditionals to determine which tiers apply.

````python
# Create a tiered sales commission calculator with rates of 5% for sales up to $1,000, 7.5% for sales between $1,001 and $3,000, and 10% for sales above $3,000
def calculate_commission(sales_amount):
    if sales_amount <= 1000:
        commission = sales_amount * 0.05
    elif sales_amount <= 3000:
        commission = 1000 * 0.05 + (sales_amount - 1000) * 0.075
    else:
        commission = 1000 * 0.05 + 2000 * 0.075 + (sales_amount - 3000) * 0.1
    return commission

# Example usage
sales = float(input("Enter sales amount: "))
commission = calculate_commission(sales)
#create and dislplay a comission breakdown that displays what is being payed out at each tier and then the total commission, base salary is $2,000, and total pay is base salary plus commission
base_salary = 2000
total_pay = base_salary + commission    
print(f"Commission Breakdown:")
if sales <= 1000:
    print(f"Tier 1: ${sales:.2f} * 5% = ${sales * 0.05:.2f}")
elif sales <= 3000:
    print(f"Tier 1: $1,000.00 * 5% = ${1000 * 0.05:.2f}")
    print(f"Tier 2: ${sales - 1000:.2f} * 7.5% = ${commission - (1000 * 0.05):.2f}")
else:
    print(f"Tier 1: $1,000.00 * 5% = ${1000 * 0.05:.2f}")
    print(f"Tier 2: $2,000.00 * 7.5% = ${2000 * 0.075:.2f}")
    print(f"Tier 3: ${sales - 3000:.2f} * 10% = ${commission - (1000 * 0.05 + 2000 * 0.075):.2f}")
print(f"Total Commission: ${commission:.2f}")
print(f"Total Pay: ${total_pay:.2f}")
````

## Problem 4: Student Grade Processor (20 points)

Create a program that processes student grades and provides analysis.

**Requirements**:

1. Create a list of at least 8 student scores (you can make them up)

2. For each score, determine the letter grade:
   - A: 90-100
   - B: 80-89
   - C: 70-79
   - D: 60-69
   - F: below 60

3. Display:
   - Each score with its letter grade
   - Class statistics:
     - Total number of students
     - Average score (rounded to 2 decimals)
     - Highest score
     - Lowest score
   - Grade distribution (count of each letter grade)
   - Percentage of students who passed (grade of D or better)

**Example Output**:
```
STUDENT GRADES
==============
Score 1: 85 (B)
Score 2: 92 (A)
Score 3: 78 (C)
...

CLASS STATISTICS
================
Number of Students: 8
Class Average: 82.50
Highest Score: 95
Lowest Score: 65

GRADE DISTRIBUTION
==================
A: 2 students
B: 3 students
C: 2 students
D: 1 student
F: 0 students

Pass Rate: 100.0%
```

````python
# Your code here
student_grades = [92, 10, 86, 99, 75, 71, 69, 88]
#calculate and display each score and its corresponding letter grade based on the following scale: A: 90-100 B: 80-89 C: 70-79 D: 60-69 F: below 60
#print a title for the grade report
print("\nGrade Report")
print("-" * 20)
for grade in student_grades:
    if grade >= 90:
        letter_grade = "A"
    elif grade >= 80:
        letter_grade = "B"
    elif grade >= 70:
        letter_grade = "C"
    elif grade >= 60:
        letter_grade = "D"
    else:
        letter_grade = "F"
    print(f"Grade: {grade}, Letter Grade: {letter_grade}")

#calculate and display the number of students, class average, highest grade, and lowest grade
num_students = len(student_grades)
class_average = sum(student_grades) / num_students
highest_grade = max(student_grades)
lowest_grade = min(student_grades)

print(f"\nClass Statistics:")
print("-" * 20)
print(f"Number of Students: {num_students}")
print(f"Class Average: {class_average:.2f}")
print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")

#create and display a new section with the number of students in each letter grade category
grade_distribution = {"A": 0, "B": 0, "C": 0, "D": 0, "F": 0}
for grade in student_grades:
    if grade >= 90:
        grade_distribution["A"] += 1
    elif grade >= 80:
        grade_distribution["B"] += 1
    elif grade >= 70:
        grade_distribution["C"] += 1
    elif grade >= 60:
        grade_distribution["D"] += 1
    else:
        grade_distribution["F"] += 1

print(f"\nGrade Distribution:")
print("-" * 20)
for letter, count in grade_distribution.items():
    print(f"{letter}: {count} students")
````

## Problem 5: Interactive Business Analyzer (15 points)

Create an interactive program that helps analyze business performance.

**Requirements**:

1. Ask the user how many months of data they want to enter (minimum 3)

2. Create two empty lists: `monthly_revenue` and `monthly_expenses`

3. Use a loop to ask the user to input revenue and expenses for each month

4. Calculate for each month:
   - Profit (revenue - expenses)
   - Profit margin percentage (profit / revenue Ã— 100)

5. Display:
   - Monthly breakdown showing revenue, expenses, profit, and margin
   - Overall summary:
     - Total revenue
     - Total expenses
     - Total profit
     - Average monthly profit
     - Best performing month (highest profit)
     - Number of profitable months

**Example Output**:
```
MONTHLY BUSINESS ANALYSIS
=========================
Month 1: Revenue: $10000, Expenses: $7000, Profit: $3000 (30.0% margin)
Month 2: Revenue: $12000, Expenses: $8000, Profit: $4000 (33.3% margin)
Month 3: Revenue: $9000, Expenses: $7500, Profit: $1500 (16.7% margin)

OVERALL SUMMARY
===============
Total Revenue: $31000
Total Expenses: $22500
Total Profit: $8500
Average Monthly Profit: $2833.33
Best Month: Month 2 ($4000 profit)
Profitable Months: 3 out of 3
```

````python
#ask the user how many months they want to calculate for
num_months = int(input("How many months do you want to calculate for? "))
monthly_revenue = []
monthly_expenses = []
for month in range(1, num_months + 1):
    revenue = float(input(f"Enter revenue for month {month}: "))
    expenses = float(input(f"Enter expenses for month {month}: "))
    monthly_revenue.append(revenue)
    monthly_expenses.append(expenses)
#print out a formatted table showing month number, revenue, expenses, and profit/loss for each month with the percentage margin in parentheses after profit 
print("\nMonthly Financial Report")
print("-" * 50)
print(f"{'Month':<6} {'Revenue':<12} {'Expenses':<12} {'Profit/Loss':<15} {'Margin (%)':<12}")
for month in range(num_months):
    profit_loss = monthly_revenue[month] - monthly_expenses[month]
    margin = (profit_loss / monthly_revenue[month] * 100) if monthly_revenue[month] != 0 else 0
    print(f"{month + 1:<6} ${monthly_revenue[month]:<11.2f} ${monthly_expenses[month]:<11.2f} ${profit_loss:<14.2f} {margin:<11.2f}%")

#calculate and display total revenue, total expenses, total profit/loss, and average monthly profit/loss as well as the best month (with the amount) and worst month (with the amount) and how many months were profitable versus unprofitable
total_revenue = sum(monthly_revenue)
total_expenses = sum(monthly_expenses)
total_profit_loss = total_revenue - total_expenses
average_monthly_profit_loss = total_profit_loss / num_months
best_month = max(range(num_months), key=lambda m: monthly_revenue[m] - monthly_expenses[m])
worst_month = min(range(num_months), key=lambda m: monthly_revenue[m] - monthly_expenses[m])
profitable_months = sum(1 for month in range(num_months) if monthly_revenue[month] - monthly_expenses[month] > 0)
unprofitable_months = num_months - profitable_months
#now print the summary
print(f"\nFinancial Summary:")
print("-" * 50)
print(f"Total Revenue: ${total_revenue:.2f}")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Total Profit/Loss: ${total_profit_loss:.2f}")
print(f"Average Monthly Profit/Loss: ${average_monthly_profit_loss:.2f}")
print(f"Best Month: Month {best_month + 1} with ${monthly_revenue[best_month] - monthly_expenses[best_month]:.2f}")
print(f"Worst Month: Month {worst_month + 1} with ${monthly_revenue[worst_month] - monthly_expenses[worst_month]:.2f}")
print(f"Profitable Months: {profitable_months}")
print(f"Unprofitable Months: {unprofitable_months}")
````
