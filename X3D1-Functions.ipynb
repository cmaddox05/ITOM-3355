## Part 2: Anatomy of a Function

### The Four Main Parts:

```python
def function_name(parameter1, parameter2):  # â† Function HEADER
    '''
    This is a DOCSTRING - explains what the function does
    '''                                      # â† DOCSTRING
    result = parameter1 + parameter2        # â† Function BODY (code)
    return result                           # â† RETURN statement
```

Let's examine each part:

### 1. Function Header (Signature)

```python
def function_name(parameter1, parameter2):
```

**Components:**
- `def` - keyword that means "define a function"
- `function_name` - describes what the function does
- `(parameter1, parameter2)` - inputs the function needs
- `:` - colon is REQUIRED!

**Parameters** are placeholders for values the function will receive.

### 2. Docstring

```python
'''
Explains what the function does.
Describes parameters and what gets returned.
'''
```

**Purpose:**
- Documents what the function does
- Helps Copilot generate correct code
- Helps other programmers (and future you!) understand your code

**Best Practice**: Always write clear docstrings!

### 3. Function Body

```python
    result = parameter1 + parameter2
```

**The actual code that does the work:**
- Must be indented (4 spaces)
- Can be multiple lines
- Uses the parameters
- Performs calculations, makes decisions, etc.

### 4. Return Statement

```python
    return result
```

**Purpose:**
- Sends a value back to whoever called the function
- Ends the function execution
- Only ONE value can be returned (but it can be a list!)
- Not all functions need to return something

## Part 3: Your First Function

### Example 1: Simple Addition Function

````python
def add_numbers(num1, num2):
    '''
    num1 and num2 are two numbers.
    Return the sum of the two numbers.
    '''
    total = num1 + num2
    return total

# Test it!
result = add_numbers(5, 3)
print(result)  # Should print 8
````

### Example 2: Business Function - Calculate Total with Tax

````python
def calculate_total_with_tax(price, tax_rate):
    '''
    price is the base price of an item.
    tax_rate is the tax rate as a decimal (e.g., 0.08 for 8%).
    Return the total price including tax.
    '''
    tax = price * tax_rate
    total = price + tax
    return total

# Test it!
final_price = calculate_total_with_tax(100, 0.08)
print(f"Final price: ${final_price}")  # Should be $108.00
````

## Part 4: Parameters vs Arguments

### Important Distinction!

**Parameters** = Variables in the function definition (placeholders)
```python
def greet(name, time_of_day):  # name and time_of_day are PARAMETERS
    return f"Good {time_of_day}, {name}!"
```

**Arguments** = Actual values passed when calling the function
```python
message = greet("Alice", "morning")  # "Alice" and "morning" are ARGUMENTS
```

**Analogy**: Parameters are like blanks in a form. Arguments are the actual values you fill in!

````python
# Example demonstrating parameters vs arguments
def multiply(x, y):  # x and y are parameters
    '''
    Return the product of x and y.
    '''
    return x * y

# Calling with different arguments
result1 = multiply(3, 4)      # 3 and 4 are arguments
result2 = multiply(10, 5)     # 10 and 5 are arguments
result3 = multiply(2.5, 2)    # 2.5 and 2 are arguments

print(result1, result2, result3)
````

## Part 6: Return vs Print

### CRITICAL Difference!

**return** = Sends a value BACK to the caller (can be used in calculations)
**print** = Shows text on screen (can't be used in calculations)

### Example of the Difference:

````python
# Function that RETURNS
def add_return(a, b):
    return a + b

# Function that PRINTS
def add_print(a, b):
    print(a + b)

# Try to use them
result1 = add_return(5, 3)
result2 = add_print(5, 3)

print(f"result1 = {result1}")  # 8 - We can use this!
print(f"result2 = {result2}")  # None - Can't use this in calculations!
````

**Key Point**: Use `return` when you want to use the result in other calculations. Use `print` only when you want to show information to the user.

## Part 7: Functions Can Call Other Functions!

### Building Blocks Approach

One function can use another function. This is how we manage complexity!

````python
%time
# Small function - calculates tax
def calculate_tax(price, tax_rate):
    '''
    Return the tax amount for the given price and tax rate.
    '''
    return price * tax_rate

# Another small function - calculates tip
def calculate_tip(price, tip_percent):
    '''
    Return the tip amount for the given price and tip percentage.
    '''
    return price * (tip_percent / 100)

# Bigger function - uses the other two!
def calculate_restaurant_total(meal_price, tax_rate, tip_percent):
    '''
    Calculate the total restaurant bill including tax and tip.
    '''
    tax = calculate_tax(meal_price, tax_rate)
    tip = calculate_tip(meal_price, tip_percent)
    total = meal_price + tax + tip
    return total

# Test it
bill = calculate_restaurant_total(50, 0.08, 20)
print(f"Total bill: ${bill:.2f}")
````

### Exercise 2: Area Calculator

````python
def rectangle_area(length, width):
    '''
    Return the area of a rectangle.
    '''
    # Your code
    area = length * width
    return area

def circle_area(radius):
    '''
    Return the area of a circle.
    Use 3.14159 for pi.
    '''
    # Your code
    pi = 3.14159
    area = pi * (radius ** 2)
    return area 

# Test
print(rectangle_area(5, 10))  # Should be 50
print(circle_area(5))          # Should be ~78.54
````

### Exercise 3: Business Functions

````python
def calculate_profit(revenue, costs):
    '''
    Return the profit (revenue minus costs).
    '''
    # Your code
    profit = revenue - costs
    return profit

def calculate_profit_margin(revenue, costs):
    '''
    Return the profit margin as a percentage.
    Formula: (profit / revenue) * 100
    '''
    # Hint: Use the calculate_profit function!
    profit = calculate_profit(revenue, costs)
    if revenue == 0:
        return 0
    return (profit / revenue * 100)


# Test
print(calculate_profit(1000, 600))         # 400
print(calculate_profit_margin(1000, 600))  # 40.0
````

### Exercise 4: Grade Calculator

````python
def average_of_three(num1, num2, num3):
    '''
    Return the average of three numbers.
    '''
    # Your code
    average = (num1 + num2 + num3) / 3
    return average

def letter_grade(score):
    '''
    score is a number from 0-100.
    Return the letter grade (A, B, C, D, or F).
    90+ = A, 80-89 = B, 70-79 = C, 60-69 = D, below 60 = F
    '''
    # Your code
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# Test
avg = average_of_three(85, 90, 88)
print(f"Average: {avg}, Grade: {letter_grade(avg)}")
````

### Challenge 1: Loan Calculator

````python
def calculate_simple_interest(principal, rate, time):
    '''
    principal is the loan amount.
    rate is the annual interest rate (as decimal, e.g., 0.05 for 5%).
    time is the number of years.
    Return the simple interest amount.
    Formula: principal * rate * time
    '''
    # Your code
    interest = principal * rate * time
    return interest

def total_loan_payment(principal, rate, time):
    '''
    Return the total amount to be paid back (principal + interest).
    Use the calculate_simple_interest function!
    '''
    # Your code
    interest = calculate_simple_interest(principal, rate, time)
    total_payment = principal + interest
    return total_payment

# Test: $10,000 loan at 5% for 3 years
print(f"Total to repay: ${total_loan_payment(10000, 0.05, 3):.2f}")
````

### Challenge 2: Employee Pay Calculator

````python
def calculate_regular_pay(hours, rate):
    '''
    Calculate regular pay for up to 40 hours.
    '''
    # Your code
    if hours <= 40:
        return hours * rate
    else:
        return 40 * rate

def calculate_overtime_pay(hours, rate):
    '''
    Calculate overtime pay (1.5x rate for hours over 40).
    If hours <= 40, return 0.
    '''
    # Your code
    if hours > 40:
        overtime_hours = hours - 40
        overtime_pay = overtime_hours * rate * 1.5
        return overtime_pay
    else:
        return 0

def calculate_total_pay(hours, rate):
    '''
    Calculate total weekly pay including overtime.
    Use both previous functions!
    '''
    # Your code
    regular_pay = calculate_regular_pay(hours, rate)
    overtime_pay = calculate_overtime_pay(hours, rate)
    total_pay = regular_pay + overtime_pay
    return total_pay

# Test: 45 hours at $20/hour
print(f"Total pay: ${calculate_total_pay(45, 20):.2f}")
````

````python
#labor efficiency calculator
def calculate_labor_efficiency(output, labor_hours):
    '''
    output is the number of units produced.
    labor_hours is the total hours of labor used.
    Return the labor efficiency (output per labor hour).
    '''
    
    if labor_hours == 0:
        return 0
    efficiency = output / labor_hours
    return efficiency

#currency converter from USD to EUR
def usd_to_eur(usd_amount, exchange_rate):
    '''
    usd_amount is the amount in US dollars.
    exchange_rate is the conversion rate from USD to EUR (e.g., 0.85).
    Return the equivalent amount in Euros.
    '''
    eur_amount = usd_amount * exchange_rate
    return eur_amount

#Payback period calculator
def calculate_payback_period(initial_investment, annual_cash_flow):
    '''
    initial_investment is the total amount invested.
    annual_cash_flow is the amount of cash flow generated each year.
    Return the payback period in years, calculated by dividing the initial investment by the annual cash flow.
    '''
    if annual_cash_flow == 0:
        return float('inf')  # Infinite payback period if no cash flow
    payback_period = initial_investment / annual_cash_flow
    return payback_period

#shipping cost calculator in EUR
def calculate_shipping_cost_EUR(weight, distance):
    '''
    weight is the weight of the package in kilograms.
    distance is the shipping distance in kilometers.
    Find the shipping cost calculated as (weight * 0.5) + (distance * 0.1).
    Convert it to EUROs using the usd_to_eur function with an exchange rate of 0.85.
    Return the shipping cost in EUROs.
    '''
    cost = (weight * 0.5) + (distance * 0.1)
    return usd_to_eur(cost, 0.85)

#10-year investment at a flat rate calculator
def calculate_investment_value(initial_investment, annual_cash_flow):
    '''
    initial_investment is the amount of money invested.
    Annual_cash_flow is the amount of cash flow generated each year.
    multiply the annual cash flow by 10 (for 10 years) and add it to the initial investment to find the total value of the investment after 10 years.
    calculate payback period using the calculate_payback_period function and return both the total value and the payback period as a tuple.
    ''' 
    total_value = initial_investment + (annual_cash_flow * 10)
    payback_period = calculate_payback_period(initial_investment, annual_cash_flow)
    return total_value, payback_period

# Test the functions
print(calculate_labor_efficiency(100, 20))  # Should be 5
print(usd_to_eur(100, 0.85))  # Should be 85.0
print(calculate_payback_period(1000, 200))  # Should be 5.0
print(calculate_shipping_cost_EUR(10, 100))  # Should be approximately 12.75
print(calculate_investment_value(1000, 200))  # Should be (3000, 5.0)
````
